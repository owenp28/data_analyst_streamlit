### Pertanyaan 1 : 

```python
# Question 1 : How does PM2.5 concentration vary by month?
df_cleaned = pd.read_csv('PRSA_Data_Wanliu_20130301-20170228.csv')

# Question 1: How does PM2.5 concentration vary by month?
df_cleaned['year_month'] = pd.to_datetime(df_cleaned[['year', 'month', 'day']]).dt.to_period('M')
monthly_avg_pm25 = df_cleaned.groupby('year_month')['PM2.5'].mean()

# Insight: Analyze monthly average PM2.5 trends.
print("\nMonthly Average PM2.5:")
print(monthly_avg_pm25)
```

### Pertanyaan 2:

```python
# Question 2: What is the correlation between PM2.5 and other pollutants?
correlation_matrix = df_cleaned[['PM2.5', 'PM10', 'CO', 'NO2', 'SO2']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Insight: Strong correlations between certain pollutants may indicate common sources
```

## Visualization & Explanatory Analysis

```python
# Visualization 1: Monthly PM2.5 Trends
plt.figure(figsize=(12, 6))
monthly_avg_pm25.plot(kind='line', title="Monthly Average PM2.5 Concentration", ylabel="PM2.5", xlabel="Month", grid=True)
plt.show()

# Visualization 2: Heatmap for Correlation
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap Between Pollutants")
plt.show()
```
